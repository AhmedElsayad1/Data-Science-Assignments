```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
sns.set(rc={'figure.figsize': [10, 10]}, font_scale=1.2)
```

```python
df = pd.read_csv('assignment/50ulke.csv')
df.head()
```

## * `Study Growth feature`

```python
sns.distplot(df['Growth'])
```

#### - The most average growth for most countries is located between 2 and 3 

## *`Study National Income feature`

```python
sns.kdeplot(df['National Income'], color='m', clip=(2000,22000) )
```

####  - most countries have a national income between 2000 to 4000
####  - the most income common between these countries is approximately 2500

## *`Study the relation between National Income and Growth`

```python
sns.jointplot(x='National Income', y='Growth', data=df)
```

#### - most of countries that have a large growth have lower national income

## *`Study the relation between National Income and contients`

```python
sns.boxplot(x="Continent", y="National Income", data=df)
```

#### - North America is the highest contient that has nationanl income
#### - other contients have a national income less than 2500
#### - Africa has least numbers of Countries that share in nationanl income

## *`Study the relation between Continent and Per person`

```python
sns.stripplot(x="Continent", y="Per person", data=df)
```

#### - Europe is the highest continent in income per person
#### - Africa and North America are the lowest between other continents
